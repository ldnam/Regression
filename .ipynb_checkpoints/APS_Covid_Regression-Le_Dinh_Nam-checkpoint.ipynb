## **(1) Data pre-processing**

```python
# Import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
```

```python
# Import dataset
covid_df = pd.read_csv('https://gist.githubusercontent.com/ldnam/079235dd43269bbd6eab3f957e032f12/raw/88b00b611a8353fb20c857bf8579409c43325337/owid-covid-2022.csv')
print(covid_df.columns)
covid_df
```

**1.** First, we need to remove unnecessary columns (columns that do not influence to the model)

```python
# Remove unnecessary columns
drop_cols = ['iso_code', 'population', 'tests_units']
covid_df = covid_df.drop(drop_cols, axis = 1)
print(covid_df.columns)
covid_df
```

**2.** Following the instructions, we will convert the 'continent' columns to seperate columns

```python
# Convert 'continent' to seperate columns
for continent in covid_df['continent'].unique():
  new_col = [1 if i == continent else 0 for i in covid_df['continent']]
  covid_df['is_' + continent.lower().replace(' ', '_')] = new_col

covid_df = covid_df.drop(['continent'], axis = 1)
covid_df
```

**3.** Also, we only use data of the first six days to predict the result of the seventh day. Then, we can compare with the real value of day 7. The next step is to flatten the data of the first six days to only one record, so each country is represented as a row in the table.

```python
# Extract day 7 data
def to_date(date):
  return '02/0' + str(date) + '/2022'

day7_df = covid_df[covid_df.date == to_date(7)][['location', 'new_cases_per_million', 'new_deaths_per_million']]
train_df = covid_df.drop(day7_df.index).drop(['date'], axis = 1)

# Linear intepolate missing value for each country
# Countries with all NaN values in a column will not be interpolated
def linear_interpolate(data): 
  return data.interpolate(limit_direction='both')

interpolate_cols = [col for col in train_df.columns if train_df[col].isna().values.any()]

for country in train_df['location'].unique():
  for col in interpolate_cols:
    train_df.loc[train_df.location == country, (col)] = linear_interpolate(train_df.loc[train_df.location == country, (col)])

# Flatten 6 days to 1 record
train_df = train_df.groupby('location', as_index=False).mean()
```

```python
day7_df
```

```python
train_df
```

**4.** We continue to drop countries with a lot of missing data.

```python
# Dropping countries with too much missing data
cols_max_nul = 7 # maximum NaN cols

null_drop_countries = []
for country in covid_df['location'].unique():
  if train_df[train_df.location == country].iloc[0].isna().sum() > cols_max_nul:
    null_drop_countries.append(country)

print('Total drop: ' + str(len(null_drop_countries)) + ' - ' + str(null_drop_countries))

# Drop in both train and day 7 data
for country in null_drop_countries:
  train_df = train_df.drop(train_df[train_df.location == country].index)
  day7_df = day7_df.drop(day7_df[day7_df.location == country].index)

train_df
```

**5.** Finally, we check the nullity of each column, and remove those having more than 40% of nullity (the threshold can be changed to make the model more accurate).

```python
# Column nullity summary (% NaN cols in all cols)
def nullity_check(df):
  nullity = {
      title: df[title].isna().sum() / df.shape[0] for title in df.columns
      }
  return nullity

nullity = nullity_check(train_df)
nullity
```

```python
# Drop cols with high nulity
null_threshold = 0.4
null_drop_cols = [col for col in nullity if nullity[col] > null_threshold]

train_df = train_df.drop(null_drop_cols, axis = 1)

print(null_drop_cols)

train_df
```

**6.** For columns that still have nullity, we will impute missing values.

```python
# Impute missing values using KNN

imputer = KNNImputer(n_neighbors = 3, weights = 'distance')

cols = train_df.columns.tolist()
cols.remove('location')

train_df.loc[:,cols] = imputer.fit_transform(train_df.loc[:,cols])
train_df
```

```python
# Create corresponding vector X and y for linear regression: min|Xw-y|

y1 = day7_df.drop(['location'], axis = 1)['new_cases_per_million']
y1.reset_index(drop=True, inplace=True)
y2 = day7_df.drop(['location'], axis = 1)['new_deaths_per_million']
y2.reset_index(drop=True, inplace=True)

X = train_df.drop(['location'], axis = 1)
X.reset_index(drop=True, inplace=True)
```

```python
print(y1)
print(y2)
```

```python
# Normalize X columns
X_cols = X.columns
X = MinMaxScaler().fit_transform(X)
X = pd.DataFrame(X, columns = X_cols)

# Insert constant column of 1
X.insert(0, 'const', pd.Series([1 for x in range(X.shape[0])]))

X
```

## **(2) Regression with standard diagnostics**

```python
# Using Statsmodels to perform linear regression and report standard diagnostics
result1 = sm.OLS(y1, X).fit()
result2 = sm.OLS(y2, X).fit()

# Prediction of the 7th day
pred1 = result1.predict(X)
pred2 = result2.predict(X)
```

```python
# Summary for 'new_cases_per_million' linear regression
print(result1.summary())

print('\n\n')
# Summary for 'new_deaths_per_million' linear regression
print(result2.summary())
```

```python
print('new_cases_per_million RSME: \t' + str(result1.mse_total**0.5))
print('new_deaths_per_million RSME: \t' + str(result2.mse_total**0.5))
```

```python
def plot_prediction(location, pred, real):
  fig = plt.gcf()
  fig.set_size_inches(15, 8)
  pred_line, = plt.plot(location, pred, label='Prediction')
  real_line, = plt.plot(location, real, label='Real')
  plt.xticks([])
  plt.legend()
  plt.show()

# Plot for 'new_cases_per_million
plot_prediction(train_df['location'].to_numpy(), pred1, y1)
print()
# Plot for 'new_deaths_per_million'
plot_prediction(train_df['location'].to_numpy(), pred2, y2)
```

## **(4) Step-wise forward regression**

```python
from matplotlib.cbook import print_cycles
# Stepwise forward regression

alpha = 0.05

def add_constant(df):
  tmp = df.copy()
  tmp.insert(0, 'const', pd.Series([1 for x in range(df.shape[0])]))
  return tmp

def forward_stepwise(X_data, y_data, ols_result):
  best_cols = []

  while True:
    unused_cols = [col for col in X_data.columns if col not in best_cols]
    next_best_col = None
    next_best_pval = alpha

    for col in unused_cols:
      tmp_cols = best_cols + [col]
      tmp_result = sm.OLS(y_data, add_constant(X_data[tmp_cols])).fit()

      col_pval = tmp_result.pvalues[col]

      if col_pval < next_best_pval:
        next_best_pval = col_pval
        next_best_col = col
      
    if next_best_col == None:
      break
    else:
      best_cols.append(next_best_col)

    new_result = sm.OLS(y_data, add_constant(X_data[best_cols])).fit()
    if max(new_result.pvalues.drop('const')) > alpha:
      best_cols.pop()
      break
      
  print(best_cols)

  return sm.OLS(y_data, add_constant(X_data[best_cols])).fit(), best_cols

# Regression for 'new_cases_per_million'
new_result1, cols_result1 = forward_stepwise(X.drop('const', axis = 1), y1, result1)
print(new_result1.summary())

print('\n\n')
# Regression for 'new_deaths_per_million'
new_result2, cols_result2 = forward_stepwise(X.drop('const', axis = 1), y2, result2)
print(new_result2.summary())
```

```python
print(new_result1.mse_total**0.5)
print(new_result2.mse_total**0.5)
```

```python
plot_prediction(train_df['location'].to_numpy(), new_result1.predict(add_constant(X[cols_result1])), y1)
print()
plot_prediction(train_df['location'].to_numpy(), new_result2.predict(add_constant(X[cols_result2])), y2)
```

## **(5) Step-wise backward regression**

```python
# Stepwise forward regression

alpha = 0.05

def backward_stepwise(X_data, y_data, ols_result):
  p_vals = ols_result.pvalues
  max_p = max(p_vals)
  new_cols = X_data.columns.tolist()

  while max_p > alpha:
    max_col = [key for key in p_vals.keys() if p_vals[key] == max_p][0]

    new_cols.remove(max_col)

    new_result = sm.OLS(y_data, add_constant(X_data[new_cols])).fit()

    p_vals = new_result.pvalues.drop('const')
    max_p = max(p_vals)
  print(new_cols)
  return sm.OLS(y_data, add_constant(X_data[new_cols])).fit(), new_cols
  
new_result1, cols_result1 = backward_stepwise(X.drop('const', axis = 1), y1, result1)
print(new_result1.summary())
new_result2, cols_result2 = backward_stepwise(X.drop('const', axis = 1), y2, result2)
print(new_result2.summary())
```

```python
print(new_result1.mse_total**0.5)
print(new_result2.mse_total**0.5)
```

```python
plot_prediction(train_df['location'].to_numpy(), new_result1.predict(add_constant(X[cols_result1])), y1)
print()
plot_prediction(train_df['location'].to_numpy(), new_result2.predict(add_constant(X[cols_result2])), y2)
```

## **(6) Discuss differences between forward and backward step-wise regression**

There are several certain differences between forward and backward step-wise regression:

### **1. Method**
The method of forward regression applies the rule that we will iteratively put the most significant variable into the model until a pre-specified stopping rule is reached or all the variables under consideration are included in the model. Indeed, by fixing a threshold and finding the p-value of each variable, we will add those with low p-values (smaller than the threshold) and run the model again and again.

In constrast, the backward regression starts with a full list of all variables and our task is to remove those with high p-values (larger than the threshold) and run the model with the new list. Then, we will do this process again until the all of the p-values are smaller than the threshold.
